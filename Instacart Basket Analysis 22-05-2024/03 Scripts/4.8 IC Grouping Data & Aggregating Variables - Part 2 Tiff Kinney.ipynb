## (Q6) Create a spending flag for each user based on the average price across all their orders.

```python
# Create new column, 'user_mean_product_price' using transform()
ords_prods_merge['user_mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check relevant columns
ords_prods_merge[['user_id', 'prices', 'user_mean_product_price']].head(500)
```

Create spending_flag column using loc() functions

```python
# 'Low spender' when user_mean_product_price less than 10
ords_prods_merge.loc[ords_prods_merge['user_mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# 'High spender' when user_mean_product_price greater than or equal to 10
ords_prods_merge.loc[ords_prods_merge['user_mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check Dataframe
ords_prods_merge[['user_id', 'user_mean_product_price', 'spending_flag']].head(100)
```

## (Q7) Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column.

```python
# Similar to the last step, start by creating a column for the median 'days_since_prior_order' grouped by user_id
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
pd.options.display.max_columns = None
```

```python
# Check dataframe
ords_prods_merge.head(100)
```

```python
# Create frequency flag (non-frequent customer if median_days_between_orders > 20)
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Create frequency flag (Regular customer if 10 < median_days_between_orders <= 20)
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 10) & (ords_prods_merge['median_days_between_orders'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Create frequency flag (Frequent customer if median_days_between_orders <= 10)
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check
ords_prods_merge[['user_id', 'median_days_between_orders', 'frequency_flag']].head(1000)
```

```python
# Check frequency of frequency_flag
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

The 5 NaN values most likely indicate that 5 customers across the whole database have only ever made 1 purchase (hence no median value for which the flags can use).
