## If-Statements with User-Defined Functions

```python
# Creating a function to assign price categories to products
def price_label(row):
    
    if row["prices"] <= 5:
        return "Low-range product"
    elif (row["prices"] > 5 and row["prices"] <= 15):
        return "Mid-range product"
    elif row["prices"] > 15:
        return "High-range product"
    else: return "More data needed"
```

```python
# Apply function to dataframe
df["price_range"] = df.apply(price_label, axis = 1)
```

```python
df["price_range"].value_counts()
```

```python
df["prices"].max()
```

## If-Statements with the loc() Function

```python
df.loc[df["prices"] > 15, "price_range_loc"] = "High-range product"
```

```python
df.loc[(df["prices"] <= 15) & (df["prices"] > 5), "price_range_loc"] = "Mid-range product"
```

```python
df.loc[df["prices"] <= 5, "price_range_loc"] = "Low-range product"
```

```python
df["price_range_loc"].value_counts()
```

## (Q4) When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# It would be a good idea to first check the frequency of values in the column order_hour_of_day
df_ords_prods_merge["order_hour_of_day"].value_counts(dropna = False)
```

Since the senior technical officer wants three categories in this new column, and without being given anything else to consider, it seems fairest to split the order_hour_of_day into equal thirds. Fortunately, this also seems appropriate given the distribution of frequencies

```python
# This new column could be made using a similar for-loop as before:
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result_3.append("Average orders")
    else:
        result_3.append("Fewest orders")
```

```python
# Check this new list
result_3
```
