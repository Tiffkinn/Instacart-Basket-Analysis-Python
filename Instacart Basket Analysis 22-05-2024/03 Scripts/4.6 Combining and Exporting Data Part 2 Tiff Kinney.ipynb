## (Q5) Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
# Need to first import the products data set
df_prods = pd.read_csv(os.path.join(path, "02 Data", "Prepared Data", "products_checked_v2.csv"), index_col = False)
```

```python
df_prods
```

```python
# Remove unnamed row
df_prods = df_prods.drop("Unnamed: 0", axis = 1)
```

```python
# Recheck
df_prods.head()
```

```python
# Using an inner merge seems to be the most ideal option since we don't want missing data in the resulting dataframe.
# An inner merge in theory should also not drop any of the rows in df_orders_products_combined
# Should remove the merge flag from the orders_products_combined dataframe
df_orders_products_combined = df_orders_products_combined.drop("_merge", axis = 1)
```

```python
# Merge
df_ords_prods_merged = df_prods.merge(df_orders_products_combined, on = "product_id", indicator = True)
```

```python
# Check output
df_ords_prods_merged
```

## (Q6) Confirm the results of the merge using the merge flag.

```python
df_ords_prods_merged["_merge"].value_counts()
```

As can be expected, since an inner join was used as the default option, all rows have information found in both dataframes
