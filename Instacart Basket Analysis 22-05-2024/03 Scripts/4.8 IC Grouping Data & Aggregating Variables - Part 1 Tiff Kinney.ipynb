## Grouping Data with pandas

```python
# Group with groupby() function
df.groupby('product_name')
```

## Aggregating Data with agg()

```python
# Average of 'order_number' grouped by department_id
df.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
# without the agg() function:
df.groupby('department_id')['order_number'].mean()
```

```python
# Aggregation with multiple statistics
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## Aggregating Data with transform()

```python
# The following single line of code does 3 things:
## Split the data into groups based on the user_id column
## Applies the transform() function on the order_number column to generate the maximum orders for each user.
## creates a max_order column to place the results of the aggregation
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# to perform a quick check of this, it would help to sort the dataframe by user_id in ascending order
ords_prods_merge.sort_values(['user_id', "order_number"], ascending=[True, True], inplace = True)
```

## Deriving Columns with loc()

Using the max_order column and the loc() function, assign loyalty labels

```python
# Loyal customers above 40 orders
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Regular customers between above 10 orders and less than or equal to 40 orders
ords_prods_merge.loc[(ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order'] <= 40), 'loyalty_flag'] = 'Regular customer'
```

```python
# New customers less than or equal to 10 orders
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = "New customer"
```

```python
# To check frequencies:
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
